## Step 2 : ETL

### Formattage des dates 

```python
df["date_added"] = pd.to_datetime(df["date_added"].astype(str).str.strip(), errors="coerce")
```

```python
print(df["date_added"])
```

### Supprimer les retours à la ligne dans les colonnes texte

```python
text_columns = df.select_dtypes(include="object").columns

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"[\r\n]+", " ", regex=True)
        .str.strip()
    )
```

### Vérification des incohérences dans les colonnes rating et duration 

```python
mask = df["rating"].str.contains("min", na=False)

df.loc[mask, "duration"] = df.loc[mask, "rating"]
df.loc[mask, "rating"] = "Not specified"
```

```python
df[df["rating"].str.contains("min", na=False)]
```

### Création de la dimension Genre


```python
# Explosion de la colonne 'listed_in' et création d'une nouvelle table 'dim_genre'

dim_genre = (
    df[["show_id", "listed_in"]]
    .assign(Genre=df["listed_in"].str.split(", "))
    .explode("Genre")
    .drop(columns="listed_in")
)
```

```python
print(dim_genre)
```

### Création de la dimension Country

```python
# Explosion de la colonne 'country' et création d'une table 'dim_countries'

dim_country = (
    df[["show_id", "country"]]
    .assign(Country=df["country"].fillna("Unknown").str.split(","))
    .explode("Country")
)

dim_country["Country"] = dim_country["Country"].str.strip()
dim_country = dim_country.drop(columns="country")
```

```python
print(dim_country)
```

### Création de la dimension Director 

```python
# Explosion de la colonne 'director' et création de la table 'dim_director'

dim_director = (
    df[["show_id", "director"]]
    .assign(Director=df["director"].fillna("Unknown").str.split(","))
    .explode("Director")
)

dim_director["Director"] = dim_director["Director"].str.strip()
dim_director = dim_director.drop(columns="director")
dim_director = dim_director[dim_director["Director"] != ""]
```

### Création de la dimension Actor

```python
# Explosion de la colonne 'cast' et création de la table 'dim_actor'

dim_actor = (
    df[["show_id", "cast"]]
    .assign(Actor=df["cast"].fillna("Unknown").str.split(","))
    .explode("Actor")
)

dim_actor["Actor"] = dim_actor["Actor"].str.strip()
dim_actor = dim_actor.drop(columns="cast")
dim_actor = dim_actor[dim_actor["Actor"] != ""]
```

### Création de la dimension Date 

```python
dim_date = pd.DataFrame({
    "Date": pd.date_range(
        start=df["date_added"].min(),
        end=df["date_added"].max(),
        freq="D"
    )
})
```

```python
dim_date["DateKey"] = dim_date["Date"].dt.strftime("%Y%m%d").astype(int)
dim_date["Year"] = dim_date["Date"].dt.year
dim_date["Quarter"] = "Q" + dim_date["Date"].dt.quarter.astype(str)
dim_date["Month"] = dim_date["Date"].dt.month
dim_date["Month_Name"] = dim_date["Date"].dt.month_name()
dim_date["Week"] = dim_date["Date"].dt.isocalendar().week.astype(int)
dim_date["Day"] = dim_date["Date"].dt.day
dim_date["Day_Name"] = dim_date["Date"].dt.day_name()
dim_date["Day_of_Week"] = dim_date["Date"].dt.weekday + 1
dim_date["Is_Weekend"] = dim_date["Day_of_Week"] >= 6
```

### Création de la dimension Type 

```python
dim_type = (
    df[["type"]]
    .drop_duplicates()
    .sort_values("type")
    .reset_index(drop=True)
)

dim_type.insert(0, "TypeKey", range(1, len(dim_type) + 1))
```

### Création de la dimension Rating

```python
dim_rating = (
    df[["rating"]]
    .drop_duplicates()
    .sort_values("rating")
    .reset_index(drop=True)
)

dim_rating.insert(0, "RatingKey", range(1, len(dim_rating) + 1))
```

### Création de la table de faits

```python
f_netflix = (
    df
    .merge(dim_type, on="type", how="left")
    .merge(dim_rating, on="rating", how="left")
    .drop(columns=[
        "listed_in",
        "country",
        "director",
        "cast",
        "type",
        "rating"
    ])
)

# Ajout de la clé étrangère vers DimDate
f_netflix = f_netflix.merge(
    dim_date[["Date", "DateKey"]],
    left_on="date_added",
    right_on="Date",
    how="left"
)

# Suppression des colonnes devenues inutiles
f_netflix = f_netflix.drop(columns=["date_added", "Date"])

# Convertir la colonne DateKey en entier
f_netflix["DateKey"] = f_netflix["DateKey"].astype("Int64")
```
